```python
import cv2
```

```python
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
from IPython.display import display
```

```python
image_path = "/Users/ayushaburman/Desktop/programming/Machine Learning/letsgrowmore/image to greyscale/dog.jpg"
```

```python
pil_image = Image.open(image_path)
```

```python
#display(pil_image)
```

# Step 1:  Read the image in RGB format

```python
image = cv2.imread(image_path)
```

# Step 2: Convert the image to grayscale

```python
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
pil_grayscale_image = Image.fromarray(grayscale_image)
```

```python
#display(pil_grayscale_image)
```

# Step 3: Invert the grayscale image (negative image)

```python
inverted_image = 255 - grayscale_image
```

```python
pil_inverted_image = Image.fromarray(inverted_image)
```

```python
#display(pil_inverted_image)
```

# Step 4: Apply Gaussian blur to the inverted image

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0, sigmaY=0)
```

```python
pil_blurred_image = Image.fromarray(blurred_image)
```

```python
#display(pil_blurred_image)
```

# Step 5: Create the pencil sketch by dividing the grayscale image by the inverted blurry image

```python
pencil_sketch = cv2.divide(grayscale_image, 255 - blurred_image, scale=256.0)
```

```python
pil_pencil_sketch = Image.fromarray(pencil_sketch)
```

```python
#display(pil_pencil_sketch)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 15))

# Plot each image in a subplot
axs[0, 0].imshow(pil_image)
axs[0, 0].axis('off')  # Hide the axes
axs[0, 0].set_title('Original')

axs[0, 1].imshow(pil_grayscale_image, cmap='gray')
axs[0, 1].axis('off')  # Hide the axes
axs[0, 1].set_title('Grayscale')

axs[1, 0].imshow(pil_inverted_image, cmap='gray')
axs[1, 0].axis('off')  # Hide the axes
axs[1, 0].set_title('Negative')

axs[1, 1].imshow(pil_pencil_sketch, cmap='gray')
axs[1, 1].axis('off')  # Hide the axes
axs[1, 1].set_title('Pencil Sketch')


# Show the subplots
plt.show()
```
